```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1, 2, 3], [2, 5, 7])
plt.title('Basic Chart')
plt.ylabel('This is Y')
plt.xlabel('This is X')
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1, 2, 3]
y1 = [2, 5, 7]
x2 = [2, 3, 4]
y2 = [2, 4, 8]
plt.plot(x1, y1, label="First")
plt.plot(x2, y2, label="Second", linewidth=3)
plt.legend()
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1, 2, 3]
y1 = [2, 5, 7]
x2 = [2, 3, 4]
y2 = [2, 4, 8]
plt.scatter(x1, y1, label="First")
plt.scatter(x2, y2, label="Second")
plt.legend()
plt.show()
```
